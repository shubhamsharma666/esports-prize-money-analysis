```python
# View rows where PercentOffline is missing in the merged dataset
# We want to see whether these rows have OfflineEarnings or TotalEarnings values that can help us fill the missing percentage
master_df[master_df["PercentOffline"].isnull()][
    ["Game", "Date", "Earnings", "ReleaseDate", "OfflineEarnings", "TotalEarnings", "PercentOffline"]
].head(25)
```

```python
# Count how many rows have TotalEarnings equal to 0
# This helps us check whether missing PercentOffline might be caused by division-by-zero situations
(master_df["TotalEarnings"] == 0).sum()
```

```python
# Show rows where PercentOffline is missing and TotalEarnings is greater than 0
# If TotalEarnings is available, we may be able to calculate PercentOffline manually
master_df[
    (master_df["PercentOffline"].isnull()) & (master_df["TotalEarnings"] > 0)
][["Game", "OfflineEarnings", "TotalEarnings", "PercentOffline"]].drop_duplicates().head(25)
```

```python
# Fill missing PercentOffline values using this formula:
# PercentOffline = OfflineEarnings / TotalEarnings
# We only do this where PercentOffline is missing and TotalEarnings is greater than 0
master_df.loc[
    (master_df["PercentOffline"].isnull()) & (master_df["TotalEarnings"] > 0),
    "PercentOffline"
] = (
    master_df["OfflineEarnings"] / master_df["TotalEarnings"]
)
```

```python
# Check how many missing PercentOffline values are left after filling
master_df["PercentOffline"].isnull().sum()
```

```python
# Rename columns to make them easier to use later in SQL and Power BI
master_df = master_df.rename(columns={
    "Date": "date",
    "Game": "game",
    "Earnings": "earnings",
    "Players": "players",
    "Tournaments": "tournaments",
    "Year": "year",
    "Month": "month",
    "ReleaseDate": "release_year",
    "Genre": "genre",
    "TotalEarnings": "total_earnings",
    "OfflineEarnings": "offline_earnings",
    "PercentOffline": "percent_offline",
    "TotalPlayers": "total_players",
    "TotalTournaments": "total_tournaments"
})
```

```python
#Step 1: Calculate market share of each game

# Create a copy of the general dataset with only the columns we need
game_share_df = general_df[["Game", "Genre", "TotalEarnings"]].copy()
```

```python
# Calculate total esports earnings across all games
total_market_earnings = game_share_df["TotalEarnings"].sum()
```

```python
# Calculate each game's share of the total market
game_share_df["market_share_pct"] = (
    game_share_df["TotalEarnings"] / total_market_earnings
) * 100
```

```python
# Sort games from highest to lowest market share
game_share_df = game_share_df.sort_values(by="market_share_pct", ascending=False)
```

```python
# Show the top 10 games by market share
game_share_df.head(10)
```

```python
#Step 2: Calculate cumulative share of top games

# Calculate the cumulative market share as we go down the ranked list of games
game_share_df["cumulative_share_pct"] = game_share_df["market_share_pct"].cumsum()
```

```python
# Show the top 10 again with cumulative share included
game_share_df.head(10)
```

```python
#Step 3: Check how concentrated the market is

# Calculate how much of the market is controlled by the top 5 games
top_5_share = game_share_df.head(5)["market_share_pct"].sum()

# Calculate how much of the market is controlled by the top 10 games
top_10_share = game_share_df.head(10)["market_share_pct"].sum()

# Display both values
top_5_share, top_10_share
```

```python
#Step 4: Create a concentration bar chart
# Plot market share of the top 10 games
top_10_share_df = game_share_df.head(10)

plt.figure(figsize=(12, 6))
plt.bar(top_10_share_df["Game"], top_10_share_df["market_share_pct"])
plt.title("Top 10 Games by Esports Market Share")
plt.xlabel("Game")
plt.ylabel("Market Share (%)")
plt.xticks(rotation=75)
plt.show()
```

```python
#Step 5: Create yearly concentration trend
#Now we will see if earnings are becoming concentrated in fewer games over time.

# Create a table of yearly earnings by game
year_game_earnings = historical_df.groupby(["Year", "Game"], as_index=False)["Earnings"].sum()
```

```python
# Show the first few rows
year_game_earnings.head()
```

```python
#Step 6: Calculate yearly market share by game

# Calculate total earnings per year
year_totals = year_game_earnings.groupby("Year", as_index=False)["Earnings"].sum()
year_totals = year_totals.rename(columns={"Earnings": "year_total_earnings"})
year_totals.head()
```

```python
# Merge yearly totals back into the game-year table
year_game_earnings = year_game_earnings.merge(year_totals, on="Year", how="left")
```

```python
# Calculate each game's market share within its year
year_game_earnings["market_share_pct"] = (
    year_game_earnings["Earnings"] / year_game_earnings["year_total_earnings"]
) * 100
```

```python
# Show the result
year_game_earnings.head()
```

```python
#Step 7: Top 5 share by year

# Rank games within each year by earnings from highest to lowest
year_game_earnings["rank_in_year"] = year_game_earnings.groupby("Year")["Earnings"].rank(
    method="first",
    ascending=False
)

```

```python

# Keep only the top 5 games in each year
top_5_by_year = year_game_earnings[year_game_earnings["rank_in_year"] <= 5]
```

```python
# Preview the top 5 games by year table
top_5_by_year.head()
```

```python

# Sum the market share of the top 5 games for each year
top_5_concentration = top_5_by_year.groupby("Year", as_index=False)["market_share_pct"].sum()
top_5_concentration = top_5_concentration.rename(columns={"market_share_pct": "top_5_share_pct"})
```

```python
# Show concentration by year
top_5_concentration.head(20)
```

```python
# Plot the top 5 games' share of total esports earnings over time
plt.figure(figsize=(12, 6))
plt.plot(top_5_concentration["Year"], top_5_concentration["top_5_share_pct"], marker="o")
plt.title("Top 5 Games Share of Total Esports Earnings by Year")
plt.xlabel("Year")
plt.ylabel("Top 5 Share (%)")
plt.grid(True)
plt.show()
```

```python
# Merge genre information into the historical dataset
# We only need Game and Genre from the general dataset for this step
historical_with_genre = historical_df.merge(
    general_df[["Game", "Genre", "ReleaseDate"]],
    on="Game",
    how="left"
)
```

```python
# Check the first few rows to confirm Genre and ReleaseDate were added
historical_with_genre.head()
```

```python
# Group by Year and Genre to calculate yearly earnings by genre
genre_year_earnings = historical_with_genre.groupby(
    ["Year", "Genre"], as_index=False
)["Earnings"].sum()
```

```python
# Show the first few rows of the genre trend table
genre_year_earnings.head(20)
```

```python
# Create a pivot table so each genre becomes a separate column
genre_pivot = genre_year_earnings.pivot(
    index="Year",
    columns="Genre",
    values="Earnings"
).fillna(0)
```

```python
# Show the pivot table
genre_pivot.head()
```

```python
# Plot yearly earnings trends for all genres
plt.figure(figsize=(14, 7))

for genre in genre_pivot.columns:
    plt.plot(genre_pivot.index, genre_pivot[genre], marker="o", label=genre)

plt.title("Yearly Esports Earnings by Genre")
plt.xlabel("Year")
plt.ylabel("Earnings")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.show()
```

```python
# Filter only recent years from 2020 onward
recent_genre_trends = genre_year_earnings[genre_year_earnings["Year"] >= 2020]
```

```python
# Group recent years by genre and sum earnings
recent_genre_summary = recent_genre_trends.groupby("Genre", as_index=False)["Earnings"].sum()
```

```python
# Sort genres from highest to lowest based on recent earnings
recent_genre_summary = recent_genre_summary.sort_values(by="Earnings", ascending=False)
```

```python
# Show recent genre leaders
recent_genre_summary
```

```python
#Now let’s do release-year analysis.
# Count how many games were released in each release year
games_by_release_year = general_df.groupby("ReleaseDate", as_index=False)["Game"].count()
games_by_release_year = games_by_release_year.rename(columns={"Game": "number_of_games"})
```

```python
# Show number of esports titles by release year
games_by_release_year.head(20)
```

```python
# Calculate total earnings by release year
earnings_by_release_year = general_df.groupby("ReleaseDate", as_index=False)["TotalEarnings"].sum()
earnings_by_release_year = earnings_by_release_year.rename(columns={"ReleaseDate": "release_year"})
```

```python
# Show total earnings by release year
earnings_by_release_year.head(20)
```

```python
# Plot total earnings by game release year
plt.figure(figsize=(12, 6))
plt.bar(earnings_by_release_year["release_year"], earnings_by_release_year["TotalEarnings"])
plt.title("Total Esports Earnings by Game Release Year")
plt.xlabel("Release Year")
plt.ylabel("Total Earnings")
plt.show()
```
